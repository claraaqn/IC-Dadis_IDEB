```python
municipios_selecionados = [2601706, 2601201, 2604106, 2612406, 2614709, 2612455, 2611606, 2607901, 2611101, 2606002]
```

```python
principais_cidades = data[data['id_municipio'].isin(municipios_selecionados)]
```

```python
principais_cidades.head()
```

```python
geral = sns.scatterplot(data=principais_cidades, x='ano', y='ideb', hue='id_municipio', style='ensino', palette='bright')
geral.figure.set_size_inches(12, 8)
```

```python
id_para_municipio = {
    2601706: 'Belo Jardim',
    2601201: 'Arcoverde',
    2604106: 'Caruaru',
    2612406: 'Sanharó',
    2614709: 'Tacaimbó',
    2612455: 'Santa Cruz',
    2611606: 'Recife',
    2607901: 'Jaboatão',
    2611101: 'Petrolina',
    2606002: 'Garanhuns'
}

principais_cidades_nome = principais_cidades.copy()
principais_cidades_nome['nome_municipio'] = principais_cidades['id_municipio'].map(id_para_municipio)
```

```python
geral_nome = sns.scatterplot(data=principais_cidades_nome, x='ano', y='ideb', hue='nome_municipio', style='ensino', palette='bright')
geral_nome.figure.set_size_inches(12, 8)
```
