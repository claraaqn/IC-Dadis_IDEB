```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./src/br_inep_ideb_municipio.csv")
```

```python
id_para_municipio = {
    2601706: 'Belo Jardim',
    2601201: 'Arcoverde',
    2604106: 'Caruaru',
}
```

### ensino médio

```python
em_comparacao = data[
    (data['id_municipio'].isin([2601706, 2601201, 2604106])) &
    (data['ensino'] == 'medio') & 
    (data['rede'] == 'estadual')
] 

em_comparacao_nome = em_comparacao.copy()
em_comparacao_nome['nome municipio'] = em_comparacao['id_municipio'].map(id_para_municipio)
```

```python
sns.lineplot(em_comparacao_nome, x='ano', y='ideb', hue='nome municipio').figure.set_size_inches(10, 6)
plt.title('Média do IDEB ao longo dos anos \n Ensino Médio')
```

### ensino fundamental 1 municipal

```python
f1_comparacao = data[
    (data['id_municipio'].isin([2601706, 2601201, 2604106])) &
    (data['ensino'] == 'fundamental') & 
    (data['anos_escolares'] == 'iniciais (1-5)') &
    (data['rede'] == 'municipal')
] 

f1_comparacao_nome = f1_comparacao.copy()
f1_comparacao_nome['nome municipio'] = f1_comparacao['id_municipio'].map(id_para_municipio)
```

```python
sns.lineplot(f1_comparacao_nome, x='ano', y='ideb', hue='nome municipio').figure.set_size_inches(10, 6) # fazer regressão linear
plt.title('Média do IDEB ao longo dos anos \n Ensino FUndamental 1')
```

```python
sns.regplot(data=f1_comparacao_nome[(f1_comparacao_nome['nome municipio']) == 'Belo Jardim'], 
            x='ano', y='ideb',
            ci = None)
plt.title("Regressão Linear Belo Jardim \n Ensino Fundamental 1")
```
